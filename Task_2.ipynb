```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/ProdigyInternship/tested.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna((df['Age'].max()), inplace = True)
df['Fare'].fillna((df['Fare'].min()), inplace = True)
df['Cabin'].fillna("Unknown", inplace = True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived', loc="upper right")
plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
plt.title('Scatter Plot of Age v/s Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived')
plt.show()
```
